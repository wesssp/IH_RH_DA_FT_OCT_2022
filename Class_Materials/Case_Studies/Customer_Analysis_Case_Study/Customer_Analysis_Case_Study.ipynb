```python
def gender_clean(x):
    if x in ["F", "Femal", "female"]:
        return "F"
    elif x in ["M", "Male"]:
        return "M"
    elif x == np.nan:
        pass
    else:
        return "F"

all_data["gender"] = list(map(gender_clean, all_data["gender"]))
```

```python
#Removing outliers using 1.5*IQR technique for all numerical columns.
x_Q1, x_Q3 = np.percentile(all_data["income"], [25,75])
all_data.info()
```

```python
all_data.iloc[all_data["monthly premium auto"].isna().index, :]
```

```python
def iqr_counter(x):
    sorted(x)
    x_Q1, x_Q3 = np.percentile(x, [25,75])
    IQR = x_Q3 - x_Q1
    lower_range = x_Q1 - (1.5 * IQR)
    higher_range =  x_Q3 + (1.5 * IQR)
    return lower_range, higher_range
lower_income, higher_income = iqr_counter(all_data["income"])
```

```python
lower_income, higher_income
```

```python
income_outliers = all_data[all_data["income"].lt(lower_income) | all_data["income"].gt(higher_income)]
```

```python
all_data["income"].drop(income_outliers.index, inplace = True)
```
