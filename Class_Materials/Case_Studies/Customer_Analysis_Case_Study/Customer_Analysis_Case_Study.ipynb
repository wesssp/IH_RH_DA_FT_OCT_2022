```python
import numpy as np
import csv
import pandas as pd
```

```python
file1 = pd.read_csv("Data/file1.csv")
file2 = pd.read_csv("Data/file2.csv")
file3 = pd.read_csv("Data/file3.csv")
print(file1.columns, file2.columns, file3.columns)
```

```python
def columns_rename(file3):
    file3.rename(columns= {'Gender':'GENDER', "State": "ST"}, inplace=True)
    return file3
file3 = columns_rename(file3)
```

```python
def load_source_data():
    file1 = pd.read_csv("Data/file1.csv")
    file2 = pd.read_csv("Data/file2.csv")
    return pd.concat([file1, file2, file3], axis = 0)
```

```python
all_data = load_source_data()
```

```python
all_data
```

```python
def lower_letters_header(all_data):
    all_data.columns = [i.lower() for i in all_data.columns]
    return all_data
all_data = lower_letters_header(all_data)
```

```python
all_data = all_data[['gender', 'education', 'st', 'customer lifetime value', 'income',
       'monthly premium auto', 'number of open complaints', 'policy type',
       'vehicle class', 'total claim amount']]
all_data.head()
```

```python
all_data["customer lifetime value"] = all_data["customer lifetime value"].apply(lambda x : x.strip("%") if isinstance(x, str) else x)
all_data.info()
all_data.head()
```

```python
all_data["customer lifetime value"] = pd.to_numeric(all_data["customer lifetime value"], errors = "coerce")
```

```python
lost_parameters = list(all_data[all_data["customer lifetime value"].isna()].index)
```

```python
#all_data.drop(lost_parameters, axis = 0, inplace = True)
ltv_mean = np.mean(all_data["customer lifetime value"])
print(ltv_mean)
all_data = pd.DataFrame(all_data)

all_data = all_data.fillna({"customer lifetime value" : round(ltv_mean)})
```

```python
def gender_clean(x):
    if x in ["F", "Femal", "female"]:
        return "F"
    elif x in ["M", "Male"]:
        return "M"
    elif x == np.nan:
        pass
    else:
        return "F"

all_data["gender"] = list(map(gender_clean, all_data["gender"]))
```

```python
all_data["st"].unique()
```

```python
def state_clean(x):
    if x in ["Washington", "WA"]:
        return "Washington"
    elif x in ["Arizona", "AZ"]:
        return "Arizona"
    elif x in ["California", "Cali"]:
        return "California"
    else:
        return x

all_data["st"] = list(map(state_clean, all_data["st"]))
```
